```python
# File to Load (Remember to Change These)
file_to_load = "../02-Homework copy/04-Pandas/Instructions/HeroesOfPymoli/Resources/purchase_data.csv"
```

```python
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
#Describe Statistical Data
purchase_data.describe()
```

```python
# Display the total number of Players
players_df = purchase_data.groupby('SN')['SN'].nunique()
players_df.count()
```

```python
# Purchasing Analysis Tool
    # Count the unique items
unique_items = purchase_data.groupby ('Item Name')['Item Name'].nunique()
unique_items.count()

    # AVG price
avg_price = purchase_data["Price"].mean()
avg_price

    # Total nuber of Purchases
total_purchases = len(purchase_data)
total_purchases

    # Total Revenue
total_revenue = purchase_data["Price"].sum()
total_revenue

# Purchasing Analysis
purchasing_analysis = pd.DataFrame(
    {"Number of Unique Items":[unique_items.count()],                                 
     "Average Purchase Price":avg_price,                       
     "Total Number of Purchases":total_purchases,                   
     "Total Revenue":total_revenue
    })

purchasing_analysis
```

```python
# Gender Demographics
    # Count of Males, Females, and Other
gender_count = purchase_data.groupby ('Gender')['SN'].nunique()
gender_count

    # Percentage of Males, Females, and Other
gender_percentage = gender_count/576
gender_percentage.round(2)

#Gender Demo
gender_demo = pd.DataFrame(
    {"Total Count":gender_count,
     "Percentage of Players":gender_percentage
    })

gender_demo
```

```python
# Purchasing Analysis (Gender)
general_purch = purchase_data.groupby('Gender')
    # Purchase Count
purchase_count = general_purch['Purchase ID'].nunique()
    # Average Purchase Price
avg_purchase = general_purch["Price"].mean()
    # Total Purchase Value
total_purchase = general_purch['Price'].sum()
    # Avg Total Purchase per Person
avg_total_purch = total_purchase/purchase_count
avg_total_purch.round(2)

# Purchasing Analysis (Gender)
gender_analysis = pd.DataFrame(
    {"Purchase Count": purchase_count,
     "Average Purchase Price":avg_purchase,
     "Total Purchase Price":total_purchase,
     "Average Total Purchase per Person":avg_total_purch      
    })
gender_analysis.index.name = None
gender_analysis.style.format({"Average Purchase Price":"${:,.2f}",
                              "Total Purchase Price":"${:,.2f}",
                              "Average Total Purchase per Person":"${:,.2f}"})

gender_analysis
```

```python
# Age Demographics
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,46]
age_range = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", ">=40"]

purchase_data["Age Group"] = pd.cut(purchase_data["Age"],bins, labels=age_range)
purchase_data

age = purchase_data.groupby('Age Group')

    # Total Count
age_count = age['SN'].nunique()
    # Percentage of Players
age_perc = (age_count/576)*100

# Age Demo
age_demo = pd.DataFrame(
    {"Total Count":age_count,
    "Percentage of Players":age_perc
    })
age_demo.style.format({"Percentage of Players":"{:,.2f}"})
age_demo
```

```python
#Purchasing Analysis (AGE)

    # Purchase Count
age_count_purch = age["Purchase ID"].count()
    # AVG Purchase Price
avg_price_purch = age["Price"].mean()
    # Total Purchase Value
total_purch_val = age["Price"].sum()
    # AVG Total Purchase per Person
avg_total_purch = total_purch_val/576

#Purchasing Analysis
purch_analysis = pd.DataFrame({"Purchase Count":age_count_purch,
                                "Average Purchase Price":avg_price_purch,
                                "Total Purchase Value":total_purch_val,
                                "Average Total Purchase per Person":avg_total_purch
                            })
purch_analysis.style.format({"Average Purchase Price": "${:,.2f}",
                            "Total Purchase Value": "${:,.2f}",
                            "Average total Purchase per Person": "${:,.2f}"
                            })
purch_analysis
```

```python
# Top Spenders
gen_spenders = purchase_data.groupby("SN")
    # Purchase Count
purch_count_spend = gen_spenders["Purchase ID"].count()
    # AVG Purchase Price
avg_count_spend = gen_spenders["Price"].mean()    
    # Total Purchase Value
total_purch_val = gen_spenders["Price"].sum()

# Top Spenders
top_spenders = pd.DataFrame({"Purchase Count": purch_count_spend,
                             "Average Purchase Price": avg_count_spend,
                             "Total Purchase Value": total_purch_val
                            })

formatting_spend = top_spenders.sort_values(["Total Purchase Value"], ascending = False).head()

formatting_spend.style.format({"Average Purchase Price": "${:,.2f}",
                            "Total Purchase Value": "${:,.2f}",
                            })

```

```python
# Most Popular Items
item_gen = purchase_data[["Item ID", "Item Name", "Price"]]
item = item_gen.groupby(["Item ID", "Item Name"])

    # Purchase Count
pop_purch_count = item["Price"].count()
    # Item Price
pop_item_price = item["Price"].sum()    
    # Total Purchase Value
total_purch_item = pop_item_price/pop_purch_count

# Most Popular Items
Most_pop_items = pd.DataFrame({"Purchase Count": pop_purch_count,
                               "Item Price": pop_item_price,
                               "Total Purchase Value": total_purch_item
                            })

formatting_top_items = Most_pop_items.sort_values("Purchase Count", ascending = False).head()

formatting_spend.style.format({"Item Price": "${:,.2f}",
                               "Total Purchase Value": "${:,.2f}",
                             })
Most_pop_items.head()
```
